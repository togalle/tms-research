```python
import os
import mne
import matplotlib.pyplot as plt

# Define the folder containing .fif files and the channel you are interested in
folder_path = './dataset-cleaned'
channel_name = 'C1'  # e.g., 'EEG 001'

# List to store the data for plotting
data_list = []
times = None

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('-epo.fif') and "rsEEG" in filename:
        file_path = os.path.join(folder_path, filename)
        
        # Read the epochs file
        epochs = mne.read_epochs(file_path, preload=True)
        
        # Pick the specific channel
        picks = mne.pick_channels(epochs.info['ch_names'], [channel_name])
        
        # Extract the data for the picked channel
        data = epochs.get_data()[:, picks[0], :]
        
        # Average across epochs to get an ERP (Event-Related Potential)
        mean_data = data.mean(axis=0)
        
        # Append the data to the list
        data_list.append(mean_data)
        
        # Get the time values
        if times is None:
            times = epochs.times

# Plot the data from each file on the same graph
plt.figure(figsize=(12, 6))

for i, data in enumerate(data_list):
    plt.plot(times, data, label=f'File {i + 1}')

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title(f'Channel: {channel_name}')
plt.legend()
plt.xlim(0.9, 1.5)  # Set the x-axis limits
plt.show()
```

```python
import mne
import matplotlib.pyplot as plt

%matplotlib widget
#plt.switch_backend("TkAgg")

brainvision_file = '../../neuroaa/datasets/raw/uz_gent_1patient/TMS-EEG-H_02_S1b_spTEP_pre.vhdr'
fif_file = 'dataset-cleaned/TMS-EEG-H_02_S1b_spTEP_pre-epo.fif'

# Load the data
raw_brainvision = mne.io.read_raw_brainvision(brainvision_file, preload=True)
raw_fif = mne.read_epochs(fif_file, preload=True)

```

```python

# Specify the channel you want to plot
channel_name = 'Cz'  # Replace with the name of the channel you are interested in
n_epochs_to_show = 2  # Change this number to show more or fewer epochs

# Set the scaling factor (in µV for EEG channels)
scaling_factor = 50e-5  # This is 20 µV; adjust as needed

# Plot the specified channel from the BrainVision file
raw_brainvision.plot(picks=channel_name, title=f'BrainVision: {channel_name}', scalings=dict(eeg=scaling_factor))

# Plot the specified channel from the FIF file
raw_fif.plot(picks=channel_name, title=f'FIF: {channel_name}', n_epochs=n_epochs_to_show, scalings={'eeg': scaling_factor})
```
