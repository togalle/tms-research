```python
import os
```

```python
def find_matching_files(directory, substrings):
    # Get the list of all files in the directory
    all_files = os.listdir(directory)
    # Filter files that contain all specified substrings
    matching_files = [f for f in all_files if all(sub in f for sub in substrings)]
    return matching_files

# Specify the directory and substrings to search for
directory = './dataset-cleaned'
substrings = ["spTEP", "pre"]

# Get the list of matching files
matching_files = find_matching_files(directory, substrings)

print(matching_files)
```

```python
import mne
import numpy as np
import matplotlib.pyplot as plt

def average_channel_responses(fif_files, channels):
    all_averages = []

    for fif_file in fif_files:
        # Load the data
        epochs = mne.read_epochs(fif_file, preload=True)
        
        # Pick the specified channels
        epochs = epochs.pick_channels(channels)
        
        # Compute the average across epochs for each channel
        avg_data = epochs.get_data().mean(axis=0)
        
        # Compute the average across the specified channels
        channel_avg = avg_data.mean(axis=0)
        
        # Store the result
        all_averages.append(channel_avg)

    return all_averages, epochs.times

def plot_averages(all_averages, times, channels):
    plt.figure(figsize=(10, 6))
    for avg, channel in zip(all_averages, channels):
        plt.plot(times, avg, label=f'Average response: {channel}')

    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title('Average Channel Responses')
    plt.legend()
    plt.show()

# List of FIF files
fif_files = matching_files
fif_files = [os.path.join(".", "dataset-cleaned", filename) for filename in fif_files]

# List of channels to average
channels = ['Cz', 'C1', 'C3'] 

# Compute the average responses
all_averages, times = average_channel_responses(fif_files, channels)

# Plot the average responses
plot_averages(all_averages, times, fif_files)
```

```python
def plot_averages(all_averages, times, channels, start_time=1.0, end_time=1.3):
    # Find the indices corresponding to the start and end times
    start_idx = np.where(times >= start_time)[0][0]
    end_idx = np.where(times <= end_time)[0][-1]

    # Filter times and all_averages based on these indices
    filtered_times = times[start_idx:end_idx + 1]
    filtered_averages = [avg[start_idx:end_idx + 1] for avg in all_averages]

    plt.figure(figsize=(10, 6))
    for avg, channel in zip(filtered_averages, channels):
        plt.plot(filtered_times, avg, label=f'Average response: {channel}')

    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title('Average Channel Responses (1 to 1.3 seconds)')
    plt.show()


plot_averages(all_averages, times, fif_files)
```
